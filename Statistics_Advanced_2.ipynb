Question 1: What is hypothesis testing in statistics?
- Hypothesis testing is a formal statistical method used to make inferences about a population based on sample data. It provides a structured framework to determine if there is enough evidence to support a specific claim or "hypothesis" about a population parameter.
- The primary purpose is to distinguish between real effects and random chance.
- Test involves two competing statements:
1) Null Hypothesis: The "no change" or "no effect" assumption. It represents the status quo ,We assume this is true unless we find strong evidence against it.
2) Alternative Hypothesis :The claim you are actually trying to prove.
- Step-by-Step Process:
1) Step 1: State the Hypotheses
2) Step 2: Set the Significance Level
3) Step 3: Collect Data and Calculate a Test Statistic
4) Step 4: Find the P-value
5) Step 5: Make a Decision

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The Null Hypothesis : The null hypothesis is the default assumption that nothing interesting is happening. It assumes that any observed patterns in your data are just due to random chance.
- Goal: To be tested and potentially rejected.
- Assumption: No effect, no difference, or no relationship exists.
- Math: Always contains an equality sign (=,>=,<=)
Example: "This new fertilizer has no effect on plant growth"($H_0: \text{Growth}_{\text{new}} = \text{Growth}_{\text{old}}$)
- The Alternative Hypothesis : The alternative hypothesis is the claim you are actually trying to prove. It contradicts the null hypothesis by suggesting that there is a real, non-random effect.
- Goal: To find enough evidence to support this claim.
- Assumption: There is a significant effect or relationship.
- Math: Uses inequality signs ($\neq$, $<$, or $>$).
Example: "This new fertilizer increases plant growth" ($H_a: \text{Growth}_{\text{new}} > \text{Growth}_{\text{old}}$).

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level, denoted by the Greek letter alpha (1$\alpha$), is the "line in the sand" drawn before looking at your data. It defines how much evidence you need before you are willing to abandon the null hypothesis ($H_0$)
- 1). The Role of Alpha ($\alpha$) : The significance level represents the probability of making a Type I error—the risk of rejecting the null hypothesis when it is actually true (a "false positive").
- Common levels: $\alpha = 0.05$ (5%) is the standard in most research.7 This means you are willing to accept a 5% risk of being wrong.
- Strict levels: In high-stakes fields like medicine or particle physics, $\alpha$ might be set to $0.01$ (1%) or even lower to ensure the results are extremely unlikely to be a fluke.
2) Deciding the Outcome : The significance level acts as the benchmark for your p-value (the probability calculated from your actual data). The comparison between these two numbers determines your final decision:
- If p-value $\leq \alpha$: The result is statistically significant. You reject the null hypothesis.The evidence is strong enough to support your claim.
- If p-value $> \alpha$: The result is not statistically significant. You fail to reject the null hypothesis. Any observed difference could likely be due to random chance.
3) |If you choose a...|Risk of Type I Error (False Positive)|Risk of Type II Error (False Negative)|
|-------------------|---------------------|--------------------------------|
|Lower $\alpha$| Decreases. You are very careful not to claim a discovery by mistake.|Increases. You might miss a real effect because your "bar" for evidence is too high.|
|Higher $\alpha$| Increases. You are more likely to be fooled by random noise.|Decreases. You are more likely to detect even small, real effects.|

Question 4: What are Type I and Type II errors? Give examples of each.
- In hypothesis testing, errors occur because we are making a decision about an entire population based only on a small sample. Since that sample might not perfectly represent the whole group, we can sometimes reach the wrong conclusion.
- Type I Error (False Positive): A Type I error occurs when you reject a null hypothesis ($H_0$) that is actually true.In other words, you claim there is a significant effect or difference when there isn't one.
- Also known as: A "False Alarm.
- Probability: Defined by your significance level ($\alpha$). If you set $\alpha = 0.05$, there is a 5% chance you will commit this error.
- Example: A medical test tells a patient they have a disease (rejecting the "no disease" null) when they are actually perfectly healthy.
- Type II Error (False Negative) : A Type II error occurs when you fail to reject a null hypothesis that is actually false. This means you miss a real effect or difference that exists in the population.
- Also known as: A "Miss."
- Probability: Denoted by the Greek letter beta ($\beta$). It is closely related to the Power of a test ($1 - \beta$).
- Example: A safety inspector tests a bridge and concludes it is safe (failing to reject the "bridge is safe" null) when it actually has structural cracks that make it dangerous.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
- Both the Z-test and the T-test are used to compare means and determine if a difference is "statistically significant." The choice between them depends on two main factors: how much data you have and what you know about the population.
- Differences:
1) Z-test: Uses the Standard Normal Distribution (the classic bell curve). It assumes you have enough data or information to be very certain about the "spread" (variance) of the entire population.
2) T-test: Uses the Student’s T-Distribution. This distribution is "flatter" with "heavier tails." This extra width acts as a safety cushion to account for the uncertainty that comes with smaller samples.
- Use a Z-test if: 1) Sample size is large: Generally $n \geq 30$.
- 2) Population variance ($\sigma^2$) is known: You know the exact standard deviation of the entire population (which is rare in the real world).
- 3) Data is normally distributed: Or the sample size is large enough that the Central Limit Theorem applies.
- Use a T-test if: 1) Sample size is small: Generally $n < 30$
- 2) Population variance is unknown: You only have the standard deviation of your sample (14$s$), not the whole population.
- 3) Data is approximately normal: This is especially important for very small samples.
- In modern data science and research, the T-test is actually the "workhorse."

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.




```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Set the parameters
n = 10          # Number of trials
p = 0.5         # Probability of success in each trial
size = 10000    # Number of times we repeat the experiment

# 2. Generate random numbers following the binomial distribution
# This simulates 10,000 experiments of 10 coin flips each
data = np.random.binomial(n, p, size)

# 3. Create the histogram
# We center the bins over the integers 0 through 10
plt.hist(data, bins=np.arange(n+2)-0.5, density=True, rwidth=0.8,
         color='skyblue', edgecolor='black')

# 4. Add labels and title
plt.title(f'Binomial Distribution Histogram (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency (Probability)')
plt.xticks(range(n + 1))
plt.grid(axis='y', alpha=0.3)

# 5. Save and display the result
plt.savefig('binomial_distribution.png')
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy import stats

# 1. The Dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# 2. Define Hypotheses
# H0: The population mean is 50.0 (mu = 50.0)
# Ha: The population mean is NOT 50.0 (mu != 50.0)
null_mean = 50.0
alpha = 0.05

# 3. Calculate Statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use sample std dev

# 4. Calculate Z-statistic
# Formula: Z = (x_bar - mu) / (sigma / sqrt(n))
z_score = (sample_mean - null_mean) / (sample_std / np.sqrt(n))

# 5. Calculate P-value (Two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# 6. Results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# 7. Decision
if p_value <= alpha:
    print("Conclusion: Reject the Null Hypothesis (Statistically Significant)")
else:
    print("Conclusion: Fail to Reject the Null Hypothesis (Not Statistically Significant)")
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Simulate data from a normal distribution
# We use a mean of 50 and standard deviation of 10
mu = 50
sigma = 10
n = 500  # Sample size
np.random.seed(42)  # For reproducibility
data = np.random.normal(mu, sigma, n)

# 2. Calculate the 95% Confidence Interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_level = 0.95
degrees_of_freedom = n - 1

# Calculate the margin of error using the t-distribution
# Formula: Margin of Error = t_critical * (std_dev / sqrt(n))
standard_error = sample_std / np.sqrt(n)
t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)
margin_of_error = t_critical * standard_error

# Confidence interval bounds
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# 3. Plotting the data
plt.hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)

# Highlight mean and confidence interval on the plot
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2,
            label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(lower_bound, color='blue', linestyle='dotted', linewidth=2,
            label=f'95% CI Lower: {lower_bound:.2f}')
plt.axvline(upper_bound, color='blue', linestyle='dotted', linewidth=2,
            label=f'95% CI Upper: {upper_bound:.2f}')

# Formatting
plt.title('Simulated Normal Distribution with 95% CI for the Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Save the plot
plt.savefig('confidence_interval_plot.png')

# Output numeric results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_z_scores(data):
    # 1. Calculate Mean and Standard Deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # 2. Calculate Z-scores
    # This formula centers the data at 0 and scales it by the standard deviation
    z_scores = [(x - mean) / std_dev for x in data]

    # 3. Visualization
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, color='plum', edgecolor='black', alpha=0.7)

    # Add lines for reference
    plt.axvline(0, color='red', linestyle='solid', label='Mean (Z=0)')
    plt.axvline(1, color='blue', linestyle='--', label='+1 Std Dev')
    plt.axvline(-1, color='blue', linestyle='--', label='-1 Std Dev')

    # Formatting
    plt.title('Standardized Data (Z-Scores)')
    plt.xlabel('Z-Score (Standard Deviations from Mean)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    return z_scores

# Test the function with a random dataset
sample_data = np.random.normal(100, 15, 1000)
z_results = visualize_z_scores(sample_data)

# Print the first 5 Z-scores
print(f"First 5 Z-scores: {z_results[:5]}")
```
